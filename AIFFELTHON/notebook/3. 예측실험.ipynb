```python
# 라이브러리 가져오기
import pandas as pd
import matplotlib.pyplot as plt
import os
import glob
from tqdm import tqdm
import tensorflow as tf
from tensorflow import keras
import numpy as np
import cv2
import seaborn as sns
import random
from PIL import Image
from tensorflow.keras.preprocessing.image import load_img, img_to_array
```

# 모델 불러오기 및 이미지 분류 결과 함수 작성

```python
model = tf.keras.models.load_model('model.h5')
```

```python
def show_and_predict(dirpath, filename, class_names, img_size=224):
    filepath = os.path.join(dirpath, filename)
    image = load_img(filepath, target_size=(img_size, img_size))
    plt.imshow(image)
    plt.axis('off')
    image = img_to_array(image).reshape(1, img_size, img_size, 3)
    prediction = model.predict(image)[0]

    print("이미지 분류 결과:")
    
    # 가장 높은 확률을 가진 상위 3개의 클래스를 찾기
    top_classes_indices = np.argsort(prediction)[::-1][:3]
    
    for class_index in top_classes_indices:
        class_name = class_names[class_index]
        confidence = round(prediction[class_index] * 100, 2)
        print(f"{class_name}: {confidence}%")
        
class_names = {
    0: '여드름',
    1: '아토피',
    2: '기저세포암',
    3: '양성 각화 유사 병변',
    4: '습진',
    5: '흑색종',
    6: '건선, 회반증 및 관련 질환 ',
    7: '지루성 각화 및 기타 양성 종양',
    8: '백선사상균, 칸디다증 및 기타 곰팡이 감염',
    9: '사마귀, 멜러스컴 및 기타 바이러스 감염'
}
```

# 여드름

```python
# 이미지 디렉토리 경로
img_dir_path = '/aiffel/aiffel/test'

filename = '66.jpg'

# 예측 함수 호출
show_and_predict(img_dir_path, filename, class_names)
```
